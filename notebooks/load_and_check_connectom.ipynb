## Characteristcs

```python
# Show adj. matrix
import matplotlib.pyplot as plt

matrix_g = nx.to_numpy_array(G)
plt.imshow(
    matrix_g,
    cmap="jet",
    origin="lower",
)
plt.colorbar()
```

```python
# Show matrix parameters
import numpy as np

matrix_eigs, matrix_eig_vectors = np.linalg.eig(matrix_g)
print(f"Rank of adj. matrix: {np.linalg.matrix_rank(matrix_g)}")
corr = np.corrcoef(matrix_g)
x = [el.real for el in matrix_eigs]
y = [el.imag for el in matrix_eigs]
plt.figure()
plt.title("Eig. values")
plt.scatter(x, y, s=1)
plt.xlabel("Real $\lambda$")
plt.ylabel("Imag $\lambda$")


plt.show()
plt.close()

plt.figure()
plt.title("Correlation coefficients")
plt.imshow(corr, origin="lower")
```
